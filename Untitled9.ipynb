```python
s = np.shape(ds1.ULAT)
a = np.transpose(np.tile(ds1.dz,(s[1],s[0],1)),(2,1,0))
lon = (ds1['TLONG'].values)
lat = (ds1['TLAT'].values)
z_t = ds1['z_t'].values

dzu = xr.DataArray(
    data=a,
    dims=["z_t","nlat", "nlon"],
    coords=dict(
        lon=(["nlat", "nlon"], lon),
        lat=(["nlat", "nlon"], lat),
        z_t=z_t,
    ),
    attrs=dict(
        description="DZU.",
        units="cenm",
    ),
)

dzt = xr.DataArray(
    data=a,
    dims=["z_t","nlat", "nlon"],
    coords=dict(
        lon=(["nlat", "nlon"], lon),
        lat=(["nlat", "nlon"], lat),
        z_t=z_t,
    ),
    attrs=dict(
        description="DZT.",
        units="cenm",
    ),
)
```

```python
from tqdm import tqdm
```

```python
ds1["DZT"] = dzt
ds1["DZU"] = dzu
```

```python
ds1.DZT.attrs["long_name"] = "Thickness of T cells"
ds1.DZT.attrs["units"] = "centimeter"
ds1.DZT.attrs["grid_loc"] = "3111"
ds1.DZU.attrs["long_name"] = "Thickness of U cells"
ds1.DZU.attrs["units"] = "centimeter"
ds1.DZU.attrs["grid_loc"] = "3221"
```

```python
# make sure we have the cell volumne for calculations
VOL = (ds1.DZT * ds1.DXT * ds1.DYT).compute()
KMT = ds1.KMT.compute()

for j in tqdm(range(len(KMT.nlat))):
    for i in range(len(KMT.nlon)):
        k = KMT.values[j, i].astype(int)
        VOL.values[k:, j, i] = 0.0
```

```python
help(ds1.KMT.compute)
```

```python
ds1["VOL"] = VOL
```

```python
ds1.VOL.attrs["long_name"] = "volume of T cells"
ds1.VOL.attrs["units"] = "centimeter^3"
ds1.VOL.attrs["grid_loc"] = "3111"
```

```python
b_DIA_IMPVF_TEMP = xr.Dataset()
```

```python
metrics = {
    ("X",): ["DXU", "DXT"],  # X distances
    ("Y",): ["DYU", "DYT"],  # Y distances
    ("Z",): ["DZU", "DZT"],  # Z distances
    ("X", "Y"): ["UAREA", "TAREA"],
}    
```

```python
# here we get the xgcm compatible dataset
gridxgcm1, dsxgcm1 = pop_tools.to_xgcm_grid_dataset(
    ds1,
    periodic=False,
    metrics=metrics,
    boundary={"X": "extend", "Y": "extend", "Z": "extend"},
)

for coord in ["nlat", "nlon"]:
    if coord in dsxgcm1.coords:
        dsxgcm1 = dsxgcm1.drop_vars(coord)
```

```python
dsxgcm1
```

```python
### Set up dataset to gather budget terms
```

```python
b_DIA_IMPVF_TEMP["DIA_IMPVF_TEMP"] = -(
    gridxgcm1.diff(dsxgcm1.DIA_IMPVF_TEMP * dsxgcm1.TAREA, axis="Z") / dsxgcm1.VOL
)
```

```python
b_DIA_IMPVF_TEMP.DIA_IMPVF_TEMP.isel(z_t=10).mean(dim="time").plot(cmap="RdYlBu_r")
```
